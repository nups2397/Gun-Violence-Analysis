# 03. Exploring Relationships

## Correlations

```python
gv_data.info()
```

```python
# Select only numerical columns
numerical_columns = gv_data.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Binning Latitude and Longitude so that way i can combine them into one variable, as that makes more sense
gv_data['Latitude_Bin'] = pd.cut(gv_data['Latitude'], bins=10, labels=False)
gv_data['Longitude_Bin'] = pd.cut(gv_data['Longitude'], bins=10, labels=False)
```

```python
# feature Cross to create a combined location feature so longitude and latitude can be put together
gv_data['Location_Combined'] = gv_data['Latitude_Bin'].astype(str) + '_' + gv_data['Longitude_Bin'].astype(str)
```

```python
# Create a mapping for Location_Combined
location_mapping = {location: idx for idx, location in enumerate(gv_data['Location_Combined'].unique())}
```

```python
# Map Location_Combined to numerical values
gv_data['Location_Combined_Numeric'] = gv_data['Location_Combined'].map(location_mapping)
```

```python
gv_data.rename(columns = {'Location_Combined_Numeric' : 'Location Combined'}, inplace = True)
```

```python
# Select numerical columns excluding 'Latitude', 'Longitude', 'Latitude_Bin', and 'Longitude_Bin'
selected_columns = numerical_columns.difference(['Latitude', 'Longitude', 'Latitude_Bin', 'Longitude_Bin'])
```

```python
selected_columns_2= ['Longitude','Latitude', 'Latitude_Bin', 'Longitude_Bin', 'Location_Combined', 'Location Combined']
```

```python
# Add 'Location Combined' to the selected columns
selected_columns = selected_columns.union(['Location Combined'])
```

```python
#this graph will not have the error bar on top, by doing errorbar=None
sns.barplot(x='Congressional District', y='Number Killed', data=gv_data, dodge=False, errorbar=None) 
plt.xticks(rotation=90)
plt.xticks(ticks=range(10, 60, 10), labels=range(10, 60, 10))
plt.title('Average Number Killed Per Congressional District')
plt.show()
```
